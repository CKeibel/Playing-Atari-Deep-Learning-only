# Resized and Grayscale Model

## Pong-v0

### Training

```python
training = pd.read_csv(pong + "nn_1500Examples_85Epochs.csv")
training.index = np.arange(1, len(training) + 1)

training
```

```python
max_acc = training['Accuracy'].max()
max_idx = training['Accuracy'].idxmax()

min_loss = training['Loss'].min()
min_idx = training['Loss'].idxmin()

max_acc = round(max_acc,2)
min_loss = round(min_loss,2)

plt.figure(figsize=(15,5))
plt.plot(training['Loss'], label="Fehler")
plt.plot(training['Accuracy'], label="Performanz")
plt.scatter(max_idx, max_acc, color="red")
plt.scatter(min_idx, min_loss, color="green")
plt.xlabel("Epochs", fontsize="15")
plt.ylabel("Accuracy/Loss", fontsize="15")
plt.legend(fontsize="15")
plt.show()

print(max_acc, min_loss)
```

### Rewards

```python
rewards_nn = pd.read_csv("models_in_action/01_resized_and_grayscale_imgs/rewards/Pong-v0.csv")
rewards_nn.index = np.arange(1, len(rewards_nn) + 1)

rewards_nn
```

```python
rewards_random = pd.read_csv(randoms + "Pong-v0.csv")
rewards_random.index = np.arange(1, len(rewards_random) + 1)

rewards_random
```

```python
plt.figure(figsize=(15,5))
plt.plot(rewards_nn['Rewards'], label="neuronales Netz")
plt.plot(rewards_random['Rewards'], label="zufällige Aktionen")
plt.xlabel("Episode", fontsize="15")
plt.ylabel("Score", fontsize="15")
plt.legend(fontsize="15")
plt.show()
```

```python
np.average(rewards_nn['Rewards']), np.average(rewards_random['Rewards'])
```
